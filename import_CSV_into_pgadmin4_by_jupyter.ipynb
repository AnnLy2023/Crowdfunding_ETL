```python
# Import dependencies.
from pathlib import Path
from sqlalchemy import create_engine, text, inspect
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
engine = create_engine('postgresql+psycopg2://postgres:password@localhost/csvtest_db')
```

```python
# read in category csv
df = pd.read_csv("Resources/category.csv")
df.to_sql("category",con = engine, index = False, if_exists="append")
pd.read_sql(text("Select * from category"), con = engine)
```

```python
# read in contacts csv
df = pd.read_csv("Resources/contacts.csv")
df.to_sql("contacts",con = engine, index = False, if_exists="append")
pd.read_sql(text("Select * from contacts"), con = engine)
```
