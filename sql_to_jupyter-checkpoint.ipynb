```python
import pandas as pd
from sqlalchemy import create_engine, text
import matplotlib.pyplot as plt
```

```python
engine = create_engine('postgresql://postgres:password@localhost:5432/crowdfunding_db')
```

```python
query = text("""
        select
            c.category_id,
            c.category,
            p.outcome,
            COUNT(p.outcome) as outcome_count
        from
            category as c
        join
            campaign as p on c.category_id = p.category_id
        group by
            c.category_id,
            p.outcome
        order by
            outcome desc,
            outcome_count desc;
        """)

# the query you want to run
df = pd.read_sql(query, engine) 
```

```python
df
```

```python
plt.bar(df.outcome, df.outcome_count, color="dodgerblue")
plt.title("Title Title")
plt.show()
```

```python
query = text("""
        select
            c.contact_id,
            c.first_name,
            c.last_name,
            p.company_name,
            p.goal,
            p.pledged,
            p.outcome,
            ROUND((p.pledged / p.goal)*100,2) as percent_goal
        from
            contacts as c
        join
            campaign as p on c.contact_id = p.contact_id
        order by
            percent_goal desc;
        """)

# the query you want to run
df2 = pd.read_sql(query, engine) 
```
